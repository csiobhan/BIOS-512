Use `geom_bar` to make a histogram of penguin counts by `year`.

```r
penguins |> head()
```

```r
p = ggplot(penguins, aes(x = year, color = species, fill = species)) +
    geom_bar()+
     labs(y = 'Number of Penguins')+
    facet_wrap(~ species, ncol = 1)
   
p
```

Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
options(repr.plot.width = 5, repr.plot.height = 6)
p = ggplot(penguins, aes( x = body_mass_g))+
    geom_bar(width = 40)+
    labs( x = 'Body Mass (g)', y = 'Number of penguins')+
    scale_x_continuous(labels = comma)+
    scale_y_continuous(labels = number)+
    facet_wrap(~ year, ncol = 1)+
    ggtitle("Penguin Body Mass Distribution by Year")+
    theme_light()
p

```

Is there a difference between the body mass of male and female penguins? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to compare within each species-group.**

```r
options(repr.plot.width = 5, repr.plot.height = 6)
p = ggplot(penguins, aes(x = body_mass_g, y = sex))+
    geom_point(alpha = 0.4) + facet_wrap(~species, ncol = 1, 'scale = free_x')+
    geom_jitter(width = 2, height = 0.3)+
    labs( x = 'Body Mass (g)', y = 'Sex')+
    scale_x_continuous(labels = comma)+
    ggtitle("Penguin Body Mass Distribution by Sex")+
    theme_light()
  
p
```
