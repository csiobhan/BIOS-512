```r
library('tidyverse')
```

```r
can_turbine = read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv', col_types = cols())
can_turbine |> head()
```

Each row in the table represents a single turbine.

Make a bar chart ðŸ“Š **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
filter_can = can_turbine |>
    select(objectid, province_territory, project_name)|>
    mutate(prov_sorted = province_territory %>% fct_infreq() %>%  fct_rev())
#|> group_by(province_territory) |> ftc_infreq()
#filter_can |> head()
#filter_can |> head()
p = ggplot(filter_can, aes(x = prov_sorted)) + geom_bar() + coord_flip() + labs(title="Turbine Count per Province Territory",
        y ="Count", x = "Province")
p
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
filter_man = can_turbine |>
    select(objectid, manufacturer, project_name)|>
    mutate(man_sorted = manufacturer %>% fct_lump(n = 12) |> 
               fct_infreq() |> fct_rev() |> 
               fct_relevel('Other')) #fct_infreq() %>%  fct_rev())
#|> group_by(province_territory) |> ftc_infreq()
#filter_can |> head()
#filter_man |> head()
p = ggplot(filter_man, aes(x = man_sorted)) + geom_bar() + coord_flip() + labs(title="Turbine Count per Manufacturer",
        y ="Count", x = "Manufacturer")
p
```
